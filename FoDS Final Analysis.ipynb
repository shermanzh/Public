# Vehicle Thefts in 2015
The dataset identifies the top 10 most stolen vehicles in each of the 50 states, as well as the model year of each vehicle and how many thefts occurred.

```python
for col in vtdf.columns:
    print(col)
```

```python
vtdf.columns = vtdf.columns.str.replace('/', '')
```

```python
vehicles = vtdf.MakeModel.value_counts()
vehicles
```

# Recurring Terms Observation
Since the dataset uses a column name that includes a special character, it was integral to first remove that character for easier access to the column. Looking at the vehicles listed, it seems that Ford Pickups are the most recurring stolen vehicle across every state's top 10. The Acura Integra only appears 6 times, apparently disproving the statement that I was given.

```python
integra_bools = vtdf.MakeModel == "Acura Integra"
integra_bools
```

```python
integra_bools = vtdf[integra_bools]
integra_bools
```

# Acura Integra Observation
I wanted to look deeper into more than just how many times the Acura Integra appeared in the list, so I decided to take the dataset and single out the rows with the car included. By observing the boolean series, it is clear that the vehicle has only broken top 5 of any given state only 1 time.

```python
totalthefts = vtdf.groupby('MakeModel')['Thefts'].sum()
totalthefts
```

```python
testtotal = totalthefts.to_frame()
testtotal
```

```python
totalthefts.sort_values(ascending=False)
```

# Conclusion
The resulting data observations seems to indicate that the top most stolen vehicle in the United States of America in 2015 were full size Ford pickups. This disproves the initial statement that the Acura Integra was the most stolen vehicle, given that once we delved deeper into the data, it possessed an evidently low involvement in 2015 automobile thefts. The amount of Integra thefts was only a small fraction to that of the Ford pickup.
